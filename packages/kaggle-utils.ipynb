```python
#| export
import os

# kaggle info
kaggle_run_interactive = lambda:os.getenv('KAGGLE_KERNEL_RUN_TYPE') == 'Interactive'
kaggle_run_compatition = lambda:bool(os.getenv('KAGGLE_IS_COMPETITION_RERUN'))    # If "true", code is being rerun by Kaggle for private test evaluation/scoring.
kaggle_run_safe        = lambda:not kaggle_run_interactive() and not kaggle_run_compatition()
kaggle_gpu_num         = lambda:int(os.getenv("USE_GPU", 0))


def kaggle_run_info_print():
    print(f'''
    {kaggle_run_interactive() = }
    {kaggle_run_compatition() = }
    {kaggle_run_safe()        = }
    {kaggle_gpu_num()         = }
    ''')
```

```python
kaggle_run_info_print()
```
